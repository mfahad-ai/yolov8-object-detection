🧩 Step 2: Import Libraries

```python
import cv2
from ultralytics import YOLO
import matplotlib.pyplot as plt
```

🧩 Step 3: Load YOLOv8 Pre-trained Model

```python
# Load YOLOv8 nano model (fast & suitable for lab)
model = YOLO("yolov8n.pt")
```

🧩 Step 4: Read Image from Given Path

```python
image_path = r"C:\Users\Fahad\OneDrive\Desktop\Gemini_Generated_Image_po1vhupo1vhupo1v.png"

# Read image
image = cv2.imread(image_path)

# Check if image loaded
if image is None:
    print("Error: Image not loaded. Check path.")
else:
    print("Image loaded successfully!")
```

🧩 Step 5: Run Object Detection

```python
results = model(image)
```

🧩 Step 6: Visualize Detection Results (Jupyter Friendly)

```python
# Draw bounding boxes
annotated_image = results[0].plot()

# Convert BGR to RGB for matplotlib
annotated_image = cv2.cvtColor(annotated_image, cv2.COLOR_BGR2RGB)

# Display image
plt.figure(figsize=(10, 8))
plt.imshow(annotated_image)
plt.axis("off")
plt.title("YOLOv8 Object Detection Result")
plt.show()
```

STEP 7 : OBJECT DETAILS

```python
for box in results[0].boxes:
    class_id = int(box.cls[0])
    confidence = float(box.conf[0])
    class_name = model.names[class_id]
    
    print(f"Object: {class_name}, Confidence: {confidence:.2f}")
```

```python
# Step 1: Import Libraries
from ultralytics import YOLO
import cv2
import matplotlib.pyplot as plt

# Step 2: Load YOLOv8 model
model = YOLO('yolov8n.pt')  # Pre-trained YOLOv8 nano model

# Step 3: List of image paths
image_paths = [
    r"C:\Users\Fahad\Downloads\WhatsApp Image 2025-12-21 at 09.26.54_a2e58af9.jpg",
    r"C:\Users\Fahad\Downloads\WhatsApp Image 2025-12-21 at 09.26.53_fbe07ac2.jpg",
    r"C:\Users\Fahad\Downloads\WhatsApp Image 2025-12-21 at 09.26.53_e4cda10a.jpg",
    r"C:\Users\Fahad\Downloads\WhatsApp Image 2025-12-21 at 09.26.47_a9835666.jpg"
]

# Step 4: Run object detection and visualize results
for img_path in image_paths:
    # Read image
    image = cv2.imread(img_path)
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    
    # Detect objects
    results = model(img_path)
    
    # Plot detections
    result_img = results[0].plot()
    
    # Show image with detections
    plt.figure(figsize=(10, 6))
    plt.imshow(result_img)
    plt.title(f"Detections: {img_path.split('\\')[-1]}")
    plt.axis('off')
    plt.show()
    
    # Optional: Print detection details
    print("Detected objects:")
    print(results[0].boxes)  # Boxes, confidence, class ids
    print("\n")
```
